# Feature Engineering

```python
# === Convert date columns to datetime ===
date_cols = ['Date Of Account Opening', 'Last Transaction Date', 'Payment Due Date', 'Last Credit Card Payment Date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')  # coercing invalid entries to NaT
# Find the most recent date across all date columns
latest_date = df[date_cols].max().max()
print('Most recent date in dataset:', latest_date)
print('\nThe most recent date in this dataset will serve as the most recent date in real time')
```

```python
# === Lets try to do some feature engineering ===

# Account age in years from the latest date in the dataset
df['Account Age (Years)'] = round((latest_date - df['Date Of Account Opening']).dt.days / 365, 1) # round the result to 1 decimal place

# Days since last transaction
df['Days Since Last Transaction'] = (latest_date - df['Last Transaction Date']).dt.days

# Days since last credit card payment
df['Days Since Last Credit Payment'] = (latest_date - df['Last Credit Card Payment Date']).dt.days

df
```

```python
# === Drop columns that are not needed ===
drop_cols = ['Payment Due Date', 'Last Credit Card Payment Date', 'Date Of Account Opening', 'Last Transaction Date']
df.drop(columns=drop_cols, inplace=True)
```

# Machine Learning Training

```python
# === Map loan status to binary classification ===
# Approved or Closed = 1 (for closed, the loan was initially approved and paid off. That is why i am mapping Approved and Closed as 1)
# Rejected = 0
df['Loan Status'] = df['Loan Status'].map({'Approved': 1,
                                           'Closed': 1,
                                           'Rejected': 0
                                          })
df.head()
```

```python
# === Convert string values to numbers usings label encoder ===
# List of categorical columns
cat_cols = ['Gender', 'Account Type', 'Transaction Type', 'Loan Type', 'Card Type']

# Initialize a LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to each categorical column
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

df.head(2)
```

```python
# === Scale neccessary columns using StandardScaler ===
# List of columns to scale
column_to_scale = ['Account Balance', 'Transaction Amount', 'Account Balance After Transaction', 'Loan Amount', 
                   'Interest Rate', 'Credit Limit', 'Credit Card Balance', 'Minimum Payment Due', 'Account Age (Years)']

# Initialize the scaler
scaler = StandardScaler()

# Apply Standard Scaler to each column
df[column_to_scale] = scaler.fit_transform(df[column_to_scale])

df.head(2)
```

```python
# === Split dataset into Feature and target ===
X = df.drop(columns='Loan Status', axis=1)
y = df['Loan Status']
```

```python
# === Split datasets into training and testing ===
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=26, stratify=y)
```

```python
# === Initialize models ===

models = {'LogisticRegression': LogisticRegression(random_state=26),
          'RandomForest': RandomForestClassifier(random_state=26),
          'DecisionTree': DecisionTreeClassifier(random_state=26),
          'KNeighbors': KNeighborsClassifier(),
          'SupportVectors': SVC(random_state=26),
          'NaiveBayes': GaussianNB(),
          'XGBoost': XGBClassifier()
}
```

```python
# === Train the dataset ===
results = {}
for name, model in models.items():
    print(f'<===Training {name}===>')
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print(f"\n<===Test Evaluation for {name}===>")
    test_accuracy = round(accuracy_score(y_test, y_test_pred) * 100, 2)
    test_precision = round(precision_score(y_test, y_test_pred) * 100, 2)       
    test_recall = round(recall_score(y_test, y_test_pred) * 100, 2)  
    test_f1 = f1_score(y_test, y_test_pred) * 100  

    print(f"\n<===Train Evaluation for {name}===>")
    train_accuracy = round(accuracy_score(y_train, y_train_pred) * 100, 2)
    train_precision = round(precision_score(y_train, y_train_pred) * 100, 2)       
    train_recall = round(recall_score(y_train, y_train_pred) * 100, 2)  
    train_f1 = f1_score(y_train, y_train_pred) * 100 
    results[name] = {
        'Train Accuracy': train_accuracy,
        'Train Precision': train_precision,
        'Train Recall': train_recall,
        'Train F1': train_f1,
        'Test Accuracy': test_accuracy,
        'Test Precision': test_precision,
        'Test Recall': test_recall,
        'Test F1': test_f1
    }

    print(f"{name} Performance:")
    print(f"  - Train Accuruacy : {train_accuracy:.2f}")
    print(f"  - Train Precision : {train_precision:.2f}")
    print(f"  - Train Recall: {train_recall:.2f}")
    print(f"  - Train F1  : {train_f1:.2f}")
    print(f"  - Test Accuracy  : {test_accuracy:.2f}")
    print(f"  - Test Precision  : {test_precision:.2f}")
    print(f"  - Test Recall : {test_recall:.2f}")
    print(f"  - Test F1   : {test_f1:.2f}")
    print("----------------------------------------------------------------------")

# Convert results to DataFrame
df_results = pd.DataFrame(results).T
df_results
```

#### NOTE: Since KNeighbors is the best performing model, but has low evaluation percentage, i'll try as much as posible to boost its performance by performing hyper-parameter tuning on the model

```python
# === Performing Hyperparameter tuning on KNeighbors ===
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'leaf_size': [10, 20, 30, 40, 50],
    'p': [2, 4, 6, 8, 10]
}

# Initialize KNeigbhors
knn = KNeighborsClassifier()

# GridSearchCV setup
grid = GridSearchCV(knn, param_grid, cv=5, scoring='f1_weighted', n_jobs=-1)

# Train the model
grid.fit(X_train, y_train)

# Predict and evaluate
y_test_pred = grid.predict(X_test)

print("Best Parameters:", grid.best_params_)
print("Best CV F1 Score:", grid.best_score_)
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_test_pred)) # to check other metrices

y_train_pred = grid.predict(X_train)
print("\nClassification Report on Train Set:")
print(classification_report(y_train, y_train_pred))
```

```python
# The result indicates a severe case of overfitting even after hyperparameter tuning. Model is performing at a 100% on all metrics value on the training dataset and a very low value on test dataset
# I remember that the dataset itself is imbalanced, so i'm going to apply SMOTE to the dataset
```

```python
# === Lets apply SMOTE because of inbalanced data ===
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Train the model
grid.fit(X_resampled, y_resampled)

# Predict and evaluate
y_test_pred = grid.predict(X_test)

print("Best Parameters:", grid.best_params_)
print("Best CV F1 Score:", grid.best_score_)
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_test_pred)) # to check other metrices

y_train_pred = grid.predict(X_resampled)
print("\nClassification Report on Train Set:")
print(classification_report(y_resampled, y_train_pred))
```

```python
# Result is still showing signs of overfitting even after applying SMOTE to the Hyperparameter tuning
# Let me try hyperparameter tuning on the second best performing model, which is Logistic regression and also apply SMOTE for imbalanced dataset
```

```python
# === Performing Hyperparameter tuning on Logistic Regression and also apply SMOTE ===
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'C': [0.001, 0.01, 0.1, 1, 10],
    'solver': ['saga'],  
    'l1_ratio': [0, 0.25, 0.5, 0.75, 1] 
}

# Initialize model
logreg = LogisticRegression(max_iter=10000)

# Apply smote to take care of imbalanced data
smote = SMOTE(random_state=26)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# GridSearchCV
grid = GridSearchCV(logreg, param_grid, cv=5, scoring='f1_weighted', n_jobs=-1, error_score='raise')

# Fit the model
grid.fit(X_resampled, y_resampled)

# Predict and evaluate
y_test_pred = grid.predict(X_test)

print("Best Parameters:", grid.best_params_)
print("Best CV F1 Score:", grid.best_score_)
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_test_pred))

y_train_pred = grid.predict(X_resampled)
print("\nClassification Report on Train Set:")
print(classification_report(y_resampled, y_train_pred))
```

```python
# This time, there is no clear sign of overfitting, but the metrics values are to low for me. I am not satisfied with it.
# Lets try stacking different models together to see if there would be an improvement in the models performance
```

```python
# === 1. Apply SMOTE ===
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# === 2. Define base models ===
base_learners = [
    ('lr', LogisticRegression(max_iter=1000, C=1.0, solver='saga')),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svc', SVC(probability=True, kernel='rbf', C=1.0, gamma='scale')),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
]

# === 3. Define final estimator ===
final_estimator = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# === 4. Define stacking classifier ===
stack_model = StackingClassifier(
    estimators=base_learners,
    final_estimator=final_estimator,
    cv=5,
    n_jobs=-1,
    passthrough=False 
)

# === 5. Train the stacked model ===
stack_model.fit(X_resampled, y_resampled)

# === 6. Predict and evaluate ===
y_pred = stack_model.predict(X_test)
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred))

y_train_pred = stack_model.predict(X_resampled)
print("\nClassification Report on Train Set:")
print(classification_report(y_resampled, y_train_pred))
```

```python
# --- 1. Evaluate on TRAINING data ---
y_train_pred = stack_model.predict(X_resampled)

print("\nTraining Set Evaluation:")
print(f"Accuracy:  {accuracy_score(y_resampled, y_train_pred):.4f}")
print(f"Precision: {precision_score(y_resampled, y_train_pred):.4f}")
print(f"Recall:    {recall_score(y_resampled, y_train_pred):.4f}")
print(f"F1 Score:  {f1_score(y_resampled, y_train_pred):.4f}")

# --- 2. Evaluate on TESTING data ---
y_test_pred = stack_model.predict(X_test)

print("\nTesting Set Evaluation:")
print(f"Accuracy:  {accuracy_score(y_test, y_test_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_test_pred):.4f}")
print(f"Recall:    {recall_score(y_test, y_test_pred):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_test_pred):.4f}")
```

```python
# Still noticing signs of overfitting. 
# i'll try to perform hyperparameter tunning on the base learner model, then work with its best estimators
```

```python
# === 1. Define hyperparameter grids for each base learner ===
param_grids = {
    'lr': {
        'C': [0.01, 0.1, 1.0, 10],
        'penalty': ['l1', 'l2'],
        'solver': ['saga']
    },
    'rf': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2]
    },
    'knn': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance']
    },
    'dt': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    },
    'svc': {
        'C': [0.1, 1, 10],
        'gamma': ['scale', 'auto'],
        'kernel': ['rbf']
    },
    'xgb': {
        'n_estimators': [50, 100],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.7, 1.0]
    }
}

# === 2. Tune each base learner using GridSearchCV ===
print("Tuning base learners using GridSearchCV...")

# Logistic Regression
lr = LogisticRegression(max_iter=1000, class_weight='balanced')
lr_search = GridSearchCV(lr, param_grids['lr'], cv=3, scoring='f1', n_jobs=-1)
lr_search.fit(X_train, y_train)
best_lr = lr_search.best_estimator_

# Random Forest
rf = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_search = GridSearchCV(rf, param_grids['rf'], cv=3, scoring='f1', n_jobs=-1)
rf_search.fit(X_train, y_train)
best_rf = rf_search.best_estimator_

# KNN
knn = KNeighborsClassifier()
knn_search = GridSearchCV(knn, param_grids['knn'], cv=3, scoring='f1', n_jobs=-1)
knn_search.fit(X_train, y_train)
best_knn = knn_search.best_estimator_

# Decision Tree
dt = DecisionTreeClassifier(random_state=42, class_weight='balanced')
dt_search = GridSearchCV(dt, param_grids['dt'], cv=3, scoring='f1', n_jobs=-1)
dt_search.fit(X_train, y_train)
best_dt = dt_search.best_estimator_

# SVC
svc = SVC(probability=True, class_weight='balanced')
svc_search = GridSearchCV(svc, param_grids['svc'], cv=3, scoring='f1', n_jobs=-1)
svc_search.fit(X_train, y_train)
best_svc = svc_search.best_estimator_

# XGBoost (handle imbalance via scale_pos_weight)
scale_pos_weight = np.sum(y_train == 0) / np.sum(y_train == 1)
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, scale_pos_weight=scale_pos_weight)
xgb_search = GridSearchCV(xgb, param_grids['xgb'], cv=3, scoring='f1', n_jobs=-1)
xgb_search.fit(X_train, y_train)
best_xgb = xgb_search.best_estimator_
```

```python
# === 3. Define tuned base learners for stacking ===
base_learners = [
    ('lr', best_lr),
    #('rf', best_rf),
    ('knn', best_knn),
    #('dt', best_dt),
    #('svc', best_svc),
    ('xgb', best_xgb)
]

# === 4. Define final estimator (you can also tune this if needed) ===
final_estimator = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# === 5. Create and train stacking classifier ===
stack_model = StackingClassifier(
    estimators=base_learners,
    final_estimator=final_estimator,
    cv=5,
    n_jobs=-1,
    passthrough=True
)

print("\nFitting stacked model...")
stack_model.fit(X_train, y_train)

# === 6. Evaluate ===
y_train_pred = stack_model.predict(X_train)
y_test_pred = stack_model.predict(X_test)

print("\nClassification Report on Train Set:")
print(classification_report(y_train, y_train_pred))

print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_test_pred))
```

```python
y_train_pred = stack_model.predict(X_train)

print("\nTraining Set Evaluation:")
print(f"Accuracy:  {accuracy_score(y_train, y_train_pred):.4f}")
print(f"Precision: {precision_score(y_train, y_train_pred):.4f}")
print(f"Recall:    {recall_score(y_train, y_train_pred):.4f}")
print(f"F1 Score:  {f1_score(y_train, y_train_pred):.4f}")

y_test_pred = stack_model.predict(X_test)

print("\nTesting Set Evaluation:")
print(f"Accuracy:  {accuracy_score(y_test, y_test_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_test_pred):.4f}")
print(f"Recall:    {recall_score(y_test, y_test_pred):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_test_pred):.4f}")
```

```python
# Step 1: Get original + meta-feature names
original_feature_names = X_train.columns if hasattr(X_train, 'columns') else [f"feat_{i}" for i in range(X_train.shape[1])]
meta_feature_names = [name for name, _ in base_learners]

# Combine them as feature input to the final estimator
all_features = list(original_feature_names) + meta_feature_names

# Step 2: Get feature importances from final estimator
final_model = stack_model.final_estimator_
importances = final_model.feature_importances_

# Step 3: Plot
plt.figure(figsize=(10, 6))
sorted_idx = np.argsort(importances)[::-1]
sorted_features = [all_features[i] for i in sorted_idx]
sorted_importances = importances[sorted_idx]

plt.barh(sorted_features, sorted_importances, color='skyblue')
plt.xlabel('Feature Importance')
plt.title('Feature Importances from Final Estimator (Stacked Model)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
